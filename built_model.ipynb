```python
import pandas as pd
```

```python
data = pd.read_csv('clean_data.csv')
data = data.drop(columns='Unnamed: 0')
data['DATETIMEDATA'] = pd.to_datetime(data['DATETIMEDATA'])
data
```

```python
from pycaret.regression import *
setup101 = setup(data, target='PM25', session_id = 123, date_features=['DATETIMEDATA'], normalize=True)
best_model = compare_models()
tuned = tune_model(best_model)
bagged = ensemble_model(tuned , n_estimators=20)
```

```python
save_model(bagged, 'model_lightgbm_pipeline')
load_model('model_lightgbm_pipeline')
predict_model(bagged)
```
