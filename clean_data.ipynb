```python
import pandas as pd
```

```python
data = pd.read_csv('nst.csv')
data = data.drop(columns='Unnamed: 0')
```

```python
data = data[data.isnull().sum(axis=1) < 3]
cl_null = ['DATETIMEDATA','PM25', 'PM10', 'O3', 'CO', 'NO2', 'SO2', 'WS', 'TEMP', 'RH', 'WD']
df2 = data[cl_null].fillna(data[cl_null].mean().round(2))
df2.to_csv("clean_data.csv", index=True)
df2
```
